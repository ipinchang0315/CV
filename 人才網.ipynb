```
import requests
from bs4 import BeautifulSoup
import openpyxl
workbook=openpyxl.Workbook()
workbook.create_sheet()
sheet = workbook.worksheets[0]
sheet2 = workbook.worksheets[1]

list2=[]
list1=[]
homeurl= 'https://tjn.moe.edu.tw/jobshow.php?pg=1' 

html = requests.get(homeurl)
html.encoding = 'utf-8'  
html = html.text
soup = BeautifulSoup(html,'html.parser')
total_page = soup.find(id='pages')

pages=total_page.text[0:20].find('/')
t_page = total_page.text[pages+1:pages+3]
print("共" + t_page + "頁")

for i in range(int(t_page)):
  homeurl= 'https://tjn.moe.edu.tw/jobshow.php?pg=' + str(i+1)
  print(homeurl)
  html = requests.get(homeurl)
  html.encoding = 'utf-8'  
  html = html.text
  
 # print(html)
  soup = BeautifulSoup(html,'html.parser')
  #print(soup.findAll('sampleTable').tbody.findAll('tr'))
 # print(soup.findAll('table')[0].findAll('tr') )

#  for row in soup.findAll('sampleTable')[0].tbody.findAll('tr'):
  for row in soup.findAll('tbody')[0].findAll('tr'):
    #print(row.findAll('td'))
    first_column = row.findAll('td')[0].text 
    two_column = row.findAll('td')[1].text.strip()
    third_column = row.findAll('td')[2].text.strip()
    four_column = row.findAll('td')[3].text
    five_column = row.findAll('td')[4].text.strip()
    six_column = row.findAll('td')[5].text.strip()
    url= 'https://tjn.moe.edu.tw/' + row.findAll('a')[0]["href"]
    if six_column>='2021-05-01':
      listdata=[ two_column, third_column, five_column, url]
      list1.append(listdata)
      sheet.append(listdata)
      print(two_column, third_column, five_column,six_column, url)
```
